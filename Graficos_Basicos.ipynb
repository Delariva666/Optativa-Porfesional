```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Datos de Ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
# Grafico de lineas
plt.figure(figsize=(10,4))
plt.plot(x,y)
plt.title("Grafico de lineas")
plt.xlabel('X')
plt.ylabel('sin(x)')
plt.show
```

```python
# Grafico de barras
Categorias = ['A','B','C', 'D']
valores =[3,7,2,5]
plt.figure(figsize=(10,4))
plt.bar(Categorias, valores)
plt.title("Grafico de barras")
plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.show()
```

```python
#Grafico de dispersion
x = np.random.rand(50)
y = np.random.rand(50)
plt.figure(figsize=(10,4))
plt.scatter(x, y)
plt.title("Grafico de dispersion ")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#histograma
datos = np.random.randn(1000)
x = np.random.rand(50)
y = np.random.rand(50)
plt.figure(figsize=(10,4))
plt.hist(datos, bins=30)
plt.title("histograma ")
plt.xlabel('valores ')
plt.ylabel('frecuencia')
plt.show()
```
